```python
products = tc.SFrame('~/data/amazon_baby.sframe')
```

```python
selected_words = ['awesome', 'great', 'fantastic', 'amazing', 'love', 'horrible', 'bad', 'terrible', 'awful', 'wow', 'hate']
```

```python
def awesome_count(dict):
    if words in dict:
        return dict[words]
    return 0
        
```

```python
products['word_count'] = tc.text_analytics.count_words(products['review'])
```

```python
for words in selected_words:
    products[words]=products['word_count'].apply(awesome_count)
```

```python
products = products[products['rating']!= 3]
products['sentiment'] = products['rating'] >= 4
```

```python
train_data,test_data = products.random_split(.8, seed=0)
```

```python
selected_words_model=tc.logistic_classifier.create(train_data,target='sentiment',features=selected_words)
```

```python
selected_words_model
```

```python
selected_words_model.evaluate(test_data)
```
